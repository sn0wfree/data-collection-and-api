```python
target_list=['PRC','RET','MAP_5',
 'MAP_10',
 'MAP_20',
 'MAP_30',
'MAP_50',
 'MAP_60',
 'MAP_100',
 'MAP_200',
  u'realMAP_5', u'realMAP_10', u'realMAP_20', u'realMAP_30', u'realMAP_50', u'realMAP_60', u'realMAP_100',
       u'realMAP_200']
#for target in target_list:
def GroupChunk(df,chunk,target):
    rows =df.shape[0]
    a_dict_vwrt={}
    a_dict_ewrt={}
    chunk_size =rows/chunk
    #print 'chunk: %d and chunk size: %d '%(chunk,chunk_size)
    #print 'length : %d'%rows
    for i in xrange(chunk):
        if i+1<chunk:
            #print (i*chunk_size,(i+1)*chunk_size)
            vwrt,ewrt=CalulateVWRTandEWRT(df.iloc[i*chunk_size:(i+1)*chunk_size,],target)
        else:
            #print (i*chunk_size,-1)
            vwrt,ewrt=CalulateVWRTandEWRT(df.iloc[i*chunk_size:,],target)
        a_dict_vwrt.update({'Group_%d'%i:vwrt})
        a_dict_ewrt.update({'Group_%d'%i:ewrt})
    return (a_dict_vwrt,a_dict_ewrt)    
    
    
def CalulateVWRTandEWRT(df,variable):
    if sum(df['MarketValue']) !=0:
        vwrt =sum(df[variable]*df['MarketValue'])/sum(df['MarketValue'])
        ewrt=df[variable].mean()
    else:
        vwrt=0
        ewrt=0
    return   vwrt,ewrt  

def GroupbyAndconcat(df,RankSymbol,target,GroupSymbol='date'): 
    #RankSymbol ="GoogleSearch"
    locals()['vwrt_%s'%target]=[]
    locals()['ewrt_%s'%target]=[]
    #df = df[[GroupSymbol,target,RankSymbol]].dropna()
    for date,date_df in df.groupby(GroupSymbol):
        #print date_df.dropna()
        temp= date_df.sort_values(by =RankSymbol,ascending=True)
        
        a_dict_vwrt,a_dict_ewrt=GroupChunk(temp,10,target)
        
        #a =dfgroup[dfgroup['date']==20040105][['date','TICKER','GoogleSearch']].dropna()
        locals()['vwrt_%s'%target].append(pd.DataFrame(a_dict_vwrt,index=[date]))
        locals()['ewrt_%s'%target].append(pd.DataFrame(a_dict_ewrt,index=[date]))
        #temp.shape[0]
        #locals()['%s_vwrt'%target],locals()['%s_ewrt'%target]=CalulateVWRTandEWRT(temp,target)
    vwrt =pd.concat(locals()['vwrt_%s'%target]).reset_index().sort_values(by='index')
    vwrt.columns =['date' if x=='index' else x for x in list(vwrt.columns)]
    ewrt =pd.concat(locals()['ewrt_%s'%target]).reset_index().sort_values(by='index')
    ewrt.columns =['date' if x=='index' else x for x in list(ewrt.columns)]
    return vwrt,ewrt
    
                                       
```

```python
def GroupChunk(df,chunk,target):
    rows =df.shape[0]
    a_dict={}
    chunk_size =rows/chunk
    #print 'chunk: %d and chunk size: %d '%(chunk,chunk_size)
    #print 'length : %d'%rows
    for i in xrange(chunk):
        if i+1<chunk:
            #print (i*chunk_size,(i+1)*chunk_size)
            mean_of_df=df.iloc[i*chunk_size:(i+1)*chunk_size,][target].mean()
        else:
            #print (i*chunk_size,-1)
            mean_of_df=df.iloc[i*chunk_size:,][target].mean()
        a_dict.update({i:mean_of_df})
    return a_dict
        
```
