```python
def outlierfilterout(df,level=0.99):
    for d in dfcolumn:
        series =df[d]
        low=float(series.quantile([1-level]))
        high =float(series.quantile([level]))
        series
        #high =float(prc[('Group7','PRC')].quantile([0.99]))
        #prc[('Group7','PRC')] =
        t= [high if p >high else p  for p in series]
        t2 =[ low if p <low else p  for p in t]
        df[d]= t2
    return df
def cal(bgvwrtPRC):
    if (u'Group0','PRC') not in bgvwrtPRC.columns:
        if not isinstance(list(bgvwrtPRC.columns)[0],tuple):
            bgvwrtPRC.columns=pd.MultiIndex.from_tuples(zip(bgvwrtPRC.columns,['PRC']*bgvwrtPRC.shape[1]),names=['Group', 'Variable'])
        else:
            bgvwrtPRC.columns=pd.MultiIndex.from_tuples(zip(dict(bgvwrtPRC.columns).keys(),['PRC']*bgvwrtPRC.shape[1]),names=['Group', 'Variable'])      
    else:
        pass
    ma_list = [5,10, 20,30,50, 60,100,200]
    if ('TBill1Mo','PRC') not in bgvwrtPRC.columns:
        bgvwrtPRC =bgvwrtPRC.join(otherfactor)
    else:
        pass
    bgvwrtPRC[(u'HighMinusLow','PRC')]=bgvwrtPRC[(u'Group9','PRC')]-bgvwrtPRC[('Group0','PRC')]
    #dfcolumn =[(u'HighMinusLow','PRC'),(u'Group0','PRC'), (u'Group1','PRC'), (u'Group2','PRC'), (u'Group3','PRC'), (u'Group4','PRC'),(u'Group5','PRC'), (u'Group6','PRC'), (u'Group7','PRC'), (u'Group8','PRC'), (u'Group9','PRC')]
    dfcolumn =[(u'Group0','PRC'), (u'Group1','PRC'), (u'Group2','PRC'), (u'Group3','PRC'), (u'Group4','PRC'),(u'Group5','PRC'), (u'Group6','PRC'), (u'Group7','PRC'), (u'Group8','PRC'), (u'Group9','PRC')]

    for ma in ma_list:
        for col in dfcolumn:
            bgvwrtPRC[(col[0],'RETrate')] = (bgvwrtPRC[col].diff()/bgvwrtPRC[col].shift(1))
            bgvwrtPRC[(col[0],'MA_' + str(ma))] = bgvwrtPRC[col].rolling(window=ma,center=False).mean()
            temp = bgvwrtPRC[col] -bgvwrtPRC[col[0],'MA_'+str(ma)]
            bgvwrtPRC[col[0],'buysignalforMA%s'%str(ma)]=[1 if r>0 else 0 for r in temp]
            bgvwrtPRC[col[0],'MA%stiming'%str(ma)] =bgvwrtPRC[col[0],'RETrate']*bgvwrtPRC[col[0],'buysignalforMA%s'%str(ma)]+bgvwrtPRC['TBill1Mo','PRC']*(1-bgvwrtPRC[col[0],'buysignalforMA%s'%str(ma)])
            bgvwrtPRC[col[0],'MAP%s'%str(ma)] =bgvwrtPRC[col[0],'MA%stiming'%str(ma)]-bgvwrtPRC[col[0],'RETrate']
            bgvwrtPRC[col[0],'MAP%s_dummy'%str(ma)] =[1 if n>0 else 0 for n in bgvwrtPRC[col[0],'MAP%s'%str(ma)]]
        bgvwrtPRC[(u'HighMinusLow','RETrate')]=bgvwrtPRC[(u'Group9','RETrate')]-bgvwrtPRC[('Group0','RETrate')]


        bgvwrtPRC[(u'HighMinusLow','MA_' + str(ma))]=bgvwrtPRC[(u'Group9','MA_' + str(ma))]-bgvwrtPRC[('Group0','MA_' + str(ma))]
        bgvwrtPRC[(u'HighMinusLow','MA%stiming'%str(ma))]=bgvwrtPRC[(u'Group9','MA%stiming'%str(ma))]-bgvwrtPRC[('Group0','MA%stiming'%str(ma))]
        bgvwrtPRC[(u'HighMinusLow','MAP%s'%str(ma))]=bgvwrtPRC[(u'Group9','MAP%s'%str(ma))]-bgvwrtPRC[('Group0','MAP%s'%str(ma))]
        bgvwrtPRC[u'HighMinusLow','MAP%s_dummy'%str(ma)] =[1 if n>0 else 0 for n in bgvwrtPRC[u'HighMinusLow','MAP%s'%str(ma)]]
    return bgvwrtPRC

def CalculateSharpeRatio(df,riskfree='Default'):
    df=df.dropna()
    col_name=df.columns
    if riskfree =='Default':
        if ('TBill1Mo','PRC') in df.columns:
            SPRtio =dict([(n,mean(df[n] - df[('TBill1Mo','PRC')])/df[n].std()) for n in col_name if n != ('TBill1Mo','PRC')])
        else:
            try:
                riskfree=df.xs(('TBill1Mo'),level='Group', axis=1,drop_level=False)
                SPRtio =dict([(n,mean(df[n] - riskfree)/df[n].std()) for n in col_name ])
            except Exception as e:
                print e
                raise 'Unknown riskfree asset price'
    elif isinstance(riskfree,list):
        if df.shape[0] ==len(riskfree):
            #df[('TBill1Mo','PRC')] =riskfree
            
            SPRtio =dict([(n,mean(df[n] - riskfree)/df[n].std()) for n in col_name ])
        else:
            SPRtio =dict([(n,mean(df[n] - mean(riskfree))/df[n].std()) for n in col_name ])
    elif isinstance(riskfree, (int, long, float, complex)):
        SPRtio =dict([(n,mean(df[n] - riskfree)/df[n].std()) for n in col_name ])
    else:
        raise 'unknown'
    
    return pd.Series(SPRtio)
def returndescribe(retrate,variable ='Default',riskfree='gs'):
    
    if 'MAP' in variable:
        total =float(retrate.shape[0])
        successrate = pd.Series({c:sum([1 if n >0 else 0 for n in retrate[c]])/total  for c in retrate.columns})
        return pd.concat([retrate.mean(),ttest(retrate), retrate.std(),retrate.skew(),successrate], axis=1,keys =['Mean','T-values','Std','Skew','SuccessRate'])
    elif variable=='Default':
        if riskfree =='gs':
            gsriskfree =gsewrtPRCretestpr[('TBill1Mo', 'PRC')]
            return pd.concat([retrate.mean(),ttest(retrate), retrate.std(),retrate.skew(),CalculateSharpeRatio(retrate,list(gsriskfree))], axis=1,keys =['Mean','T-values','Std','Skew','SPRatio'])
        elif riskfree =='bg':
            bgriskfree =bgewrtPRCretestpr[('TBill1Mo', 'PRC')]
            return pd.concat([retrate.mean(),ttest(retrate), retrate.std(),retrate.skew(),CalculateSharpeRatio(retrate,list(bgriskfree))], axis=1,keys =['Mean','T-values','Std','Skew','SPRatio'])
        else:
            return pd.concat([retrate.mean(),ttest(retrate), retrate.std(),retrate.skew(),CalculateSharpeRatio(retrate,list(gsewrtPRCretestpr[('TBill1Mo', 'PRC')]))], axis=1,keys =['Mean','T-values','Std','Skew','SPRatio'])
    else:
        return pd.concat([retrate.mean(),ttest(retrate), retrate.std(),retrate.skew(),CalculateSharpeRatio(retrate,list(gsewrtPRCretestpr[('TBill1Mo', 'PRC')]))], axis=1,keys =['Mean','T-values','Std','Skew','SPRatio'])

        
#gsewrtPRCretestpr[('TBill1Mo', 'PRC')]
def ttest(retrate,own=False):
    if own ==True:
        return retrate.mean()*(retrate.shape[0])**0.5/retrate.std()
    else:
        return pd.Series(scipy.stats.ttest_1samp(retrate.dropna(),0)[0],index =retrate.mean().index)
```
