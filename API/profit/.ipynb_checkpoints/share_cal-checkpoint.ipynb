```python
def CalculateMA(DataFrame,ma_list ='Default'):
    ma_list = [5,10, 20,30,50, 60,100,200]
    
    # 分别计算5日、20日、30/60/100/200日的移动平均线
    if ma_list == 'Default':
        pass
        
    elif isinstance(ma_list,list) or isinstance(ma_list,tuple) or isinstance(ma_list,set):
        ma_list =ma_list
    else:
        print 'Unknown list, use Default list:5,10,20,30,60,100,200d MA'
        ma_list = [5,10, 20,30, 60,100,200]
    for ma in ma_list:
        DataFrame['MA_' + str(ma)] = pd.rolling_mean(DataFrame['PRC'], ma)
        DataFrame = recMAP(DataFrame,('MA',ma))
    return DataFrame
        
def CalculateEMA(DataFrame,ma_list ='Default'):
    ma_list = [5,10, 20,30,50, 60,100,200]
    
    # 分别计算指数平滑移动平均线EMA
    if ma_list == 'Default':
        pass
        
    elif isinstance(ma_list,list) or isinstance(ma_list,tuple) or isinstance(ma_list,set):
        ma_list =ma_list
    else:
        print 'Unknown list, use Default list:5,10,20,30,60,100,200d EMA'
        ma_list = [5,10, 20,30, 60,100,200]
    for ma in ma_list:
        DataFrame['EMA_' + str(ma)] = pd.ewma(DataFrame['PRC'], span=ma)
        DataFrame = recMAP(DataFrame,('EMA',ma))
    return DataFrame
def recMAP(GS,ma_status):
    ma,day =ma_status
    
    
    #GS[ma]=abs(GS[ma])
    GS['temp'] =GS['PRC'] -GS[ma+'_'+str(day)]
    GS['temp2'] =[1 if number >0 else 0 for number in GS['temp']]
    GS['temp3'] = GS['RET']*GS['temp2'] +GS['TBill1Mo']*(1-GS['temp2'])
    GS[ma+'P_'+str(day)]=GS['temp3']
    GS[ma+'P_'+str(day)+'_dummy'] =[1 if 1 else 0 for n in GS[ma+'P_'+str(day)]]
    col =list(GS.columns)
    extred =[c for c in col if 'Unnamed:' in c ]
    
    for c in col:
        if 'temp' in c:
            extred.append(c)
    
    for ex in extred:
        col.pop(col.index(ex))
    return GS[col]
```

```python
def ReturnRate(onelist):
    if isinstance(onelist,list):
        pass
    else:
        onelist=list(onelist)
    end=copy.deepcopy(onelist)
    onelist.pop()
    del end[0]
    temp = [(bb-aa)/aa for (aa,bb) in zip(onelist,end)]
    temp.insert(0,"NAN")
    return temp
def DetectStatus(mas,prices):
    temp =zip(mas,prices)
    buy=[]
    for ma,price in temp:
        if price < ma:
            buy.append(1)
        else:
            buy.append(0)
    return buy
def PortfolioReturn(df,tagret_price_return,tagret_risk_free_asset_return,ma='Default List'):
    if isinstance(tagret_price_return,str):
        return_share =df[tagret_price_return]
    elif isinstance(tagret_price_return,list):
        return_share  = tagret_price_return
    else:
        print 'use default DataFrame Price Return'
        return_share=df['RET']
    if isinstance(tagret_risk_free_asset_return,str):
        return_risk_free =df[tagret_risk_free_asset_return]
    elif isinstance(tagret_risk_free_asset_return,list):
        return_risk_free  = tagret_risk_free_asset_return
    else:
        print 'use default TBill One Month Return'
        return_risk_free=df['TBill1Mo']
    if ma =='Default List':
        ma  = [5,10, 20,30, 60,100,200]
    else:
        ma=ma
    
    gg =zip(return_share,return_risk_free)
    for ma_day in ma:
        locals()['buy_%s'%ma_day]=DetectStatus(df['MA_%s'%ma_day],df['PRC'])
        temp=[]
        for b,g in zip(locals()['buy_%s'%ma_day],gg):
            temp.append(g[b])
        df['MAP_%s'%ma_day] =temp
    return df
            
```
