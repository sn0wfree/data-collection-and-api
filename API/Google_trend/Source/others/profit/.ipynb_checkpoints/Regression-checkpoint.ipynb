```python
def InitalSqliteConnection(target):

    if target != 0:
        conn = sqlite3.connect(target)
    else:
        conn = sqlite3.connect(":memory:")

    return conn
def CombineGandWithoutG(G_path,withoutG_path,ticker):
   
    tickerwthout =pd.read_csv(withoutG_path)
    e =pd.read_csv(G_path).T.reset_index()
    if 'Day' in list(e.iloc[0]):

        e.columns=[ str(t).upper() for t in e.iloc[0] ]
        e.drop([0],inplace=True)
    elif 'Day' in list(e.columns):
        e.columns=[ str(t).upper() for t in list(e.columns) ]

    else:
        pass
    e.columns=[str(i)  for i in list(e.columns)]
    e['date']=[''.join(day.split('-')) for day in e['DAY']]
    try:
        temp =e[['date',ticker]]
        temp.columns =['date','GoogleSearch']
        temp['date']=temp['date'].astype(np.int64)
    except Exception as el:
        
        print ticker,list(e.columns),e
        
        #print list(e.columns)[1],type(list(e.columns)[1])
        if ticker.capitalize() in list(e.columns):
            temp =e[['date',ticker.capitalize()]]
        elif ticker.title() in list(e.columns):
            temp =e[['date',ticker.title()]]
        elif ticker.upper() in list(e.columns):
            temp =e[['date',ticker.upper()]]
        elif ticker.lower() in list(e.columns):
            temp =e[['date',ticker.lower()]]
        else:
            error_columns =list(e.columns)
            error_columns.remove(['date'])
            error_columns.remove(['DAY'])
            if len(error_columns) == 1:
                if error_columns[0].upper() == ticker:
                    temp =e[['date',error_columns[0]]]
                    
                else:
                    pass
                temp.columns =['date',ticker]
            else:
                raise 'Error'
        temp =e[['date',ticker]]
        temp.columns =['date','GoogleSearch']
        temp['date']=temp['date'].astype(np.int64)
    
   
    return pd.merge(tickerwthout,temp,on='date',how='outer')
def Diff2DateFrame(df,n=1):
    return df.diff(n)[n:]
def Log2DateFrame(df):
    return log(df)
def TestADFandProcessOther(df,printout=False,othertype="Diff",process=[]):
    df=df.dropna()
    (adf,p_value)=adftest(df,False)
    

    if p_value>=0.05:
        #print p_value
        if othertype=="Diff":
            process.append("Diff")
            (adf,p_value,process) = TestADFandProcessOther(df.diff(1)[1:],printout,"Diff",process)
        else:
            process.append("Log")
            (adf,p_value,process) = TestADFandProcessOther(log(df),printout,"Log",process)
    else:
        pass
    return (adf,p_value,process)
def DetectADFtestandReturnVarlist(df,Need_test_list):
    var_list=[]
    for var in Need_test_list:
        adf_result =TestADFandProcessOther(df[var],process=[])
        if len(adf_result[2])==0:
            var_list.append(var)
        else:
            df['%sDiff%s'%(var,len(adf_result[2]))]=Diff2DateFrame(df[var],len(adf_result[2]))
            var_list.append('%sDiff%s'%(var,len(adf_result[2])))

    #print var,adf_result[2]
    #print adf_result
    #price_adf,price_adf_p=adftest(df['PRC'].diff(1)[1:],True)
    #google_adf,google_adf_p=adftest(df['GoogleSearch'],True)
    #adf_result =TestADFandProcessOther(df['GoogleSearch'],process=[])
    #print adf_result[2] 
    return df,var_list
def OlsTestProcess(GS,col_list,key):
    target_coeff_list={}
    target_result_list={}
    for ticker,dft in GS.groupby('TICKER'):
        #print df.columns,col_list
        df =dft[col_list]
        if df.shape[0]<10:
            pass
        else:
            #
            try:
                df,col =ChangeColumns(df,col_list)
                #print df.columns,col_list
                df,col =DetectADFtestandReturnVarlist(df,col)
                OLSresult =sm.ols(formula='~'.join(col[0:2])+'+'+"+".join(col[2:]), data=df).fit()

                target_result_list[ticker]=OLSresult
                target_coeff_list[ticker]={"%spvalue"%key:OLSresult.pvalues[1],"%scoef"%key:OLSresult.params[1]}
                #return (key,result,coeff)
            except Exception as e:
                print e,ticker
                pass 
    
    return target_result_list,target_coeff_list


def OlsTestProcess2(GS,col_list,key):
    target_coeff_list={}
    target_result_list={}
    for ticker,df in GS.groupby('TICKER'):
        df =df[col_list]
        #
        #print df.columns
        if df.shape[0]<10:
            pass
        else:
            try:
                df,col_list =ChangeColumns(df,col_list)
                df,GoogleSearch =DetectADFtestandReturnVarlist(df,col_list)
                
                #GoogleSearch =col_list
                #NewsHeat=col_list[0:1]+col_list[1:2]+col_list[3:]
                #result,coeff =OlsTestProcess('GS',(GoogleSearch,df))
                OLSresult =sm.ols(formula='~'.join(GoogleSearch[0:2])+'+'+"+".join(GoogleSearch[2:]), data=df).fit()
                #target_coeff_list.update()
    
                #target_result_list.update({"%s_Result"%key:OLSresult})  
        
                #print googlesearch.pvalues[-1]
                target_result_list[ticker]=OLSresult
                target_coeff_list[ticker]={"%spvalue"%key:OLSresult.pvalues[1],"%scoef"%key:OLSresult.params[1]}
                #return (key,result,coeff)
            except Exception as e:
                print e,ticker
                pass 
    
    return target_result_list,target_coeff_list
def adftest(series_test,printout=False):
    X = series_test.values
    result = adfuller(X,maxlag=4)
    if printout !=False:
        print('ADF Statistic: %f' % result[0])
        print('p-value: %f' % result[1])
        print('Critical Values:')
        for key, value in result[4].items():
            print('\t%s: %.3f' % (key, value))
        if result[1] <0.05:
            print "Statistical Stationary"
        else:
            print "Non-Stationary"
    else:
        pass
    return (result[0],result[1])
def ReadCSVBymulti(l):
    #var_list=['date','SICCD','TICKER','RET','TBill1Mo','PRC','VOL','MarketValue',u'MA_10',u'MAP_10',u'Mkt-RF', u'SMB', u'HML', u'RF',u'NEWS_HEAT_READ_DAVG',u'GoogleSearch']
    dataForall ='/Users/sn0wfree/Documents/python_projects/data_collection/API/profit/collection/'
    df=pd.read_csv(dataForall+l)
    #key =l.split('/')[-1].split('.')[0]
    return df
def ReadMultiCsvFileandCombineByPoolfunction(list_all):
    pool =mp.Pool()
    all_data_list =pool.map(ReadCSVBymulti,list_all)
    pool.close()
    pool.join()
    return pd.concat(all_data_list)
def SetDataType(all_data_list):
    col = all_data_list.columns
    all_data_list.dropna(inplace =True)
    if 'MarketValue' in col:
        all_data_list['MarketValue']=abs(all_data_list['MarketValue'])
    if 'PRC' in col:
        all_data_list['PRC']=abs(all_data_list['PRC'])
        
        
    if 'NEWS_HEAT_READ_DAVG' in col:
        #all_data_list['NEWS_HEAT_READ_DAVG']  =all_data_list['NEWS_HEAT_READ_DAVG'].fillna(0)
        all_data_list['NEWS_HEAT_READ_DAVG'] =all_data_list['NEWS_HEAT_READ_DAVG'].astype(uint8)
        #all_data_list.dropna(inplace =True)
    if 'date' in col:
        all_data_list['date'] =all_data_list['date'].astype(uint32)
    if 'GoogleSearch' in col:
        all_data_list['GoogleSearch'] =all_data_list['GoogleSearch'].astype(uint8)
    if 'TICKER' in col:
        all_data_list['TICKER'] =all_data_list['TICKER'].astype('category') 
    if 'SICCD' in col:
        all_data_list['SICCD'] =all_data_list['SICCD'].astype('uint16')  
    
    if 'NEWS_HEAT_PUB_DNUMSTORIES' in col:
        all_data_list['NEWS_HEAT_PUB_DNUMSTORIES'] =all_data_list['NEWS_HEAT_PUB_DNUMSTORIES'].astype(uint8)
    if 'NEWS_HEAT_READ_DAVG' in col:
        all_data_list['NEWS_HEAT_PUB_DMAX'] =all_data_list['NEWS_HEAT_PUB_DMAX'].astype(uint8)
    if 'PERMNO' in col:
        all_data_list['PERMNO'] =all_data_list['PERMNO'].astype(uint32)
    if 'SHROUT' in col:
        all_data_list['SHROUT'] =all_data_list['SHROUT'].astype(uint64)
    if 'VOL' in col:
        all_data_list['VOL'] =all_data_list['VOL'].astype(uint64)
      
    return all_data_list
#all_information_MAP_10
def ChangeColumns(dataframe,col_list):
    df = dataframe[col_list]
    col=list(df.columns)
    if 'Mkt-RF' in col:
        col[col.index('Mkt-RF')] = 'MktRF'
        df.columns =col
    else:
        pass
    return df,col
#col_list =['MAP_10','RET','PRC','Mkt-RF','SMB','HML','RF','NEWS_HEAT_READ_DAVG','GoogleSearch']
#all_information_MAP_10 =ChangeColumns(all_information_MAP_10,col_list)
def StatisticsofRegression(GScoeff,BGcoeff):
    GSdata =pd.DataFrame(GScoeff).T
    GSalldata =GSdata.count()[1]

    BGdata =pd.DataFrame(BGcoeff).T
    BGalldata =BGdata.count()[1]

    print 'Total Samples:',len(set(GS['TICKER'])),'; Total Regression Number:',GSalldata,'\n'
    GScount=GSdata[GSdata['GSpvalue']<0.05].count()[1]

    print "Google Search Available Sample (p<5%):",GScount,'; Available Rate:',float(GScount)/GSalldata
    print "Mean of Google Search  coeff:",mean(GSdata['GScoef'][GSdata['GSpvalue']<0.05])
    GSdata['GScoef'][GSdata['GSpvalue']<0.05].plot(legend=1,colors="red",title="The Distribution of Google Search Index Coefficient on MAP")
    plt.grid(True)  
    plt.show()


    print 'Total Samples:',len(set(BG['TICKER'])),'; Total Regression Number:',BGalldata,'\n'
    BGcount=BGdata[BGdata['BGpvalue']<0.05].count()[1]
    print "Bloomberg News Heat Available Sample (p<5%):", BGcount,'; Available Rate:',float(BGcount)/BGalldata
    print "Mean of Bloomberg News Heat coeff:",mean(BGdata['BGcoef'][BGdata['BGpvalue']<0.05])
    BGdata['BGcoef'][BGdata['BGpvalue']<0.05].plot(legend=1,title="The Distribution of Bloomberg News Heat Coefficient on MAP")
    plt.grid(True)  
    plt.show()
```

```python
#Grouped Statistic
BG_conn = InitalSqliteConnection('Bloomberg_grouped.db')
GS_conn = InitalSqliteConnection('GoogleSearch_grouped.db')
otherfactor =pd.read_csv('/Users/sn0wfree/Documents/python_projects/data_collection/API/otherfactor/otherfactor.csv')[['date','TBill1Mo','MktRF' ,u'SMB', u'HML',u'GS_SMB', u'GS_HML',u'BG_SMB', u'BG_HML']]

col_name =['HighMinusLow','Group_0', 'Group_1', 'Group_2', 'Group_3', 'Group_4', 'Group_5', 'Group_6', 'Group_7', 'Group_8', 'Group_9']

def CalculateSharpeRatio(df,col_name):
    SPRtio =dict([(n,mean(df[n] - df['TBill1Mo'])/df[n].std()) for n in col_name])
    return pd.Series(SPRtio)
def CalculateSuccessRatio(GS):
    return [float(df['MAP_10_dummy'].sum())/df.shape[0] for date,df in GS.groupby('date')]
    
def Description(BG_conn,GS_conn,otherfactor,col_name,target,rt,SPRtio='SPRtio'):
    GS={}
    BG={}
    for r in rt:
        locals()['BG'+target+r ]=pd.read_sql_query('select * from %s'%(target+r),BG_conn)
        locals()['BG'+target+r ]['HighMinusLow']=locals()['BG'+target+r ]['Group_9']-locals()['BG'+target+r ]['Group_0']
        locals()['GS'+target+r ]=pd.read_sql_query('select * from %s'%(target+r),GS_conn)
        locals()['GS'+target+r ]['HighMinusLow']=locals()['GS'+target+r ]['Group_9']-locals()['GS'+target+r ]['Group_0']


        locals()['BG'+target+r ] = pd.merge(locals()['BG'+target+r ],otherfactor,on='date')
        locals()['GS'+target+r ] = pd.merge(locals()['GS'+target+r ],otherfactor,on='date')
        if SPRtio == 'SPRtio':
            GS_SPRtio=CalculateSharpeRatio(locals()['GS'+target+r ],col_name)
            BG_SPRtio=CalculateSharpeRatio(locals()['BG'+target+r ],col_name)
        elif SPRtio =='Success':
            GS_SPRtio ='s'
            BG_SPRtio ='b'
            
        else:
            GS_SPRtio=CalculateSharpeRatio(locals()['GS'+target+r ],col_name)
            BG_SPRtio=CalculateSharpeRatio(locals()['BG'+target+r ],col_name)
        #print locals()['BG'+target+r ]
       
        GS[r] = pd.concat([locals()['GS'+target+r ][col_name].mean(), locals()['GS'+target+r ][col_name].std(),locals()['GS'+target+r ][col_name].skew(),GS_SPRtio], axis=1,keys =['Mean','Std','Skew','SRatio'])
        GS[r]['T-value'] =scipy.stats.ttest_1samp(locals()['GS'+target+r ][col_name],0)[0]
        GS[r]['p-value of t'] =scipy.stats.ttest_1samp(locals()['GS'+target+r ][col_name],0)[1]
        BG[r] = pd.concat([locals()['BG'+target+r ][col_name].mean(), locals()['BG'+target+r ][col_name].std(),locals()['BG'+target+r ][col_name].skew(),BG_SPRtio], axis=1,keys =['Mean','Std','Skew','SRatio'])
        BG[r]['T-value'] =scipy.stats.ttest_1samp(locals()['BG'+target+r ][col_name],0)[0]
        BG[r]['p-value of t'] =scipy.stats.ttest_1samp(locals()['BG'+target+r ][col_name],0)[1]    
        #title ='the BoxPlot of %s Grouped by  Googel Search(%s)'%(target,r))
        GS['df_%s'%r]=locals()['GS'+target+r ][col_name+['date']]
        BG['df_%s'%r]=locals()['BG'+target+r ][col_name+['date']]
        #locals()['GB'+target+r ][col_name].plot.box(title ='the BoxPlot of %s Grouped by Bloomberg News Heat(%s)'%(target,r)))


        #GSMA_10_ewrt[col_name].mean()
        #GSMA_10_ewrt[col_name].std()
        #GSMA_10_ewrt[col_name].skew()

    #pivottableGS =pd.pivot_table(GS_ewrt,index=['GoogleSearch_level_round'],values=['MAP','MARE30','PRC','Return'],aggfunc=[np.sum,np.mean,np.std],margins=True)  
    #print GSMAP_10_ewrt.mean(),GSMAP_10_ewrt.std(),GSMAP_10_ewrt.skew()
    return BG,GS
def DescriptionMA(BG_conn,GS_conn,otherfactor,col_name,target,rt,SPRtio='SPRtio'):
    GS={}
    BG={}
    for r in rt:
        locals()['BG'+target+r ]=pd.read_sql_query('select * from %s'%(target+r),BG_conn)
        locals()['BG'+target+r ]['HighMinusLow']=locals()['BG'+target+r ]['Group_9']-locals()['BG'+target+r ]['Group_0']
        locals()['GS'+target+r ]=pd.read_sql_query('select * from %s'%(target+r),GS_conn)
        locals()['GS'+target+r ]['HighMinusLow']=locals()['GS'+target+r ]['Group_9']-locals()['GS'+target+r ]['Group_0']
        locals()['BG'+target+r ]=RETdf(locals()['BG'+target+r ])
        locals()['GS'+target+r ]=RETdf(locals()['GS'+target+r ])
        locals()['GS'+target+r ].dropna(inplace=True)
        locals()['BG'+target+r ].dropna(inplace=True)
        


        locals()['BG'+target+r ] = pd.merge(locals()['BG'+target+r ],otherfactor,on='date')
        locals()['GS'+target+r ] = pd.merge(locals()['GS'+target+r ],otherfactor,on='date')
        if SPRtio == 'SPRtio':
            GS_SPRtio=CalculateSharpeRatio(locals()['GS'+target+r ],col_name)
            BG_SPRtio=CalculateSharpeRatio(locals()['BG'+target+r ],col_name)
        elif SPRtio =='Success':
            GS_SPRtio ='s'
            BG_SPRtio ='b'
            
        else:
            GS_SPRtio=CalculateSharpeRatio(locals()['GS'+target+r ],col_name)
            BG_SPRtio=CalculateSharpeRatioCalculateSharpeRatio(locals()['BG'+target+r ],col_name)
        #print locals()['BG'+target+r ]
       
        GS[r] = pd.concat([locals()['GS'+target+r ][col_name].mean(), locals()['GS'+target+r ][col_name].std(),locals()['GS'+target+r ][col_name].skew(),GS_SPRtio], axis=1,keys =['Mean','Std','Skew','SRatio'])
        GS[r]['T-value'] =scipy.stats.ttest_1samp(locals()['GS'+target+r ][col_name],0)[0]
        GS[r]['p-value of t'] =scipy.stats.ttest_1samp(locals()['GS'+target+r ][col_name],0)[1]
        BG[r] = pd.concat([locals()['BG'+target+r ][col_name].mean(), locals()['BG'+target+r ][col_name].std(),locals()['BG'+target+r ][col_name].skew(),BG_SPRtio], axis=1,keys =['Mean','Std','Skew','SRatio'])
        BG[r]['T-value'] =scipy.stats.ttest_1samp(locals()['BG'+target+r ][col_name],0)[0]
        BG[r]['p-value of t'] =scipy.stats.ttest_1samp(locals()['BG'+target+r ][col_name],0)[1]    
        #title ='the BoxPlot of %s Grouped by  Googel Search(%s)'%(target,r))
        GS['df_%s'%r]=locals()['GS'+target+r ][col_name+['date']]
        BG['df_%s'%r]=locals()['BG'+target+r ][col_name+['date']]
        #locals()['GB'+target+r ][col_name].plot.box(title ='the BoxPlot of %s Grouped by Bloomberg News Heat(%s)'%(target,r)))


        #GSMA_10_ewrt[col_name].mean()
        #GSMA_10_ewrt[col_name].std()
        #GSMA_10_ewrt[col_name].skew()

    #pivottableGS =pd.pivot_table(GS_ewrt,index=['GoogleSearch_level_round'],values=['MAP','MARE30','PRC','Return'],aggfunc=[np.sum,np.mean,np.std],margins=True)  
    #print GSMAP_10_ewrt.mean(),GSMAP_10_ewrt.std(),GSMAP_10_ewrt.skew()
    return BG,GS
def RETdf(df):
    for col in df.columns:
        if col != 'date':
            df[col]=df[col].diff()
    return df
```

```python
def ReturnADJR2andFpvalue(GStestCAPM):
    adjr={}
    for group,result in GStestCAPM.items():
        adjr[group]=(result.rsquared_adj,result.f_pvalue)
    return pd.DataFrame(adjr,index=['Adj_R2','f_pvalue']).T
#dir(BGtestCAPM['Group_0'])
```

```python
def ReturnPTandParams(GStestFF3,Model_name):
    params={}
    tvalue={}
    pvalue={}
    for key,value in GStestFF3.items():
        params.update({key :value.params})
        tvalue.update({key :value.tvalues})
        pvalue.update({key :value.pvalues})
    params,pvalue,tvalue =pd.DataFrame(params).T,pd.DataFrame(pvalue).T,pd.DataFrame(tvalue).T 
    return pd.concat([RenameColumns(params,'the (%s)Coeff of'%Model_name),RenameColumns(tvalue,'T value of'),RenameColumns(pvalue,'pvalue of the tvalue of'),RenameColumns(ReturnADJR2andFpvalue(GStestFF3),0)],axis=1)



#print BGtestFF3['HighMinusLow'].summary()

pd.concat([ReturnPTandParams(BGtestFF3,'FF3'),ReturnPTandParams(BGtestCAPM,'CAPM')],axis=1).to_csv('BGFF3testResult.csv')
pd.concat([ReturnPTandParams(GStestFF3,'FF3'),ReturnPTandParams(GStestCAPM,'CAPM')],axis=1).to_csv('GSFF3testResult.csv')
```
