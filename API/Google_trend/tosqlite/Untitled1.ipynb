```python
def scanCSVfile(target_file = 'default'):
    if target_file == 'default':
        locals_file_path = os.path.split(os.path.realpath(__file__))[0]
    else:
        locals_file_path = os.path.split(os.path.realpath(target_file))[0]
    csvfiles =[ file for file in os.listdir(locals_file_path) if '.csv' in file]
    return csvfiles
```

```python
csvfiles=scanCSVfile('test.py')
csvfiles
```

```python
temp={}
for csvfile in csvfiles:
    googlesearch = pd.read_csv(csvfile).T
    names= googlesearch.iloc[0].tolist()
    googlesearch.columns =names
    df =googlesearch.drop(googlesearch.index[:1])
    df.index.name='Date'

    for name in names:
        daf=df.pop(name)
        daf =daf.reset_index()
        daf.columns=['Date','%s'%name]
        temp.update({name:daf})
```
