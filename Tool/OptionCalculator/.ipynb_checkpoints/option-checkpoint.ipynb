```python

import pandas as pd
def calculateprofit(op,extend):
	Exercisable=op['正股价格']*(extend+1)>op['行权价']
	Profit=(op['正股价格']*(extend+1)-op['行权价'])*op['Option_Type']*Exercisable-op['Option_Price']
	return Profit,Profit/op['Option_Price']
	
	

if __name__ == '__main__':

	"""
	上海5800**深圳030***是认购权证
	上海5809**深圳038***是认沽权证
	"""
	csvfile='0700认购权证20170918.csv'
	op=pd.read_csv(csvfile)[[ '代码', '名称', '正股代码', '正股简称', '到期日',
       '权证最新价', '行权价', '兑换比率(倍)', '行权比率', '换股成本', '正股价格']]
	op.dropna(inplace=True)
	op['Option_Price']=op['权证最新价']*(1/op['兑换比率(倍)'])

	op['Option_Type']=[1]*op.shape[0]#call=1;put=-1

	#-------------
	extend=0
	
	#op['ReturnRate as stock increased %s'%str(extend)]=calculateprofit(op,extend)
	#print op['ReturnRate as stock increased %s'%str(extend)]
	for i in xrange(-100,101,1):
		
		op['Profit as stock increased %s%%'%str(i/100.0*100)],op['ReturnRate as stock increased %s%%'%str(i/100.0*100)]=calculateprofit(op,i/100.0)
	print op
    
```
