```python
param =['donor'
,'status'
,'intent'
,'year'
,'crssectorname'
,'flow_class'
,'recipient_name'
,'recipient_iso2'
,'recipient_iso3']
r2 = requests.get(
        'http://china.aiddata.org/aggregates/projects?get=%s'%(','.join(param)))
print ','.join(param)

recipent =pd.DataFrame(eval(r2.text.replace('null','"NULL"')))
```

```python
yearandrecipent_based=[]
for year,year_df in recipent.groupby('year'):
    yearandrecipent_based.extend([(year,recipients,recipients_df.usd_current.sum()) for recipients,recipients_df in year_df.groupby('recipient_name')])
         
pd.DataFrame(yearandrecipent_based,columns=['year','recipient_name','totalamount'])      
```
