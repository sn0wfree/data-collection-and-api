```python
import requests,sqlite3
import requests_cache
import pandas as pd



def InitalSqliteConnection(target):
    if target != 0:
        conn = sqlite3.connect(target)
    else:
        conn = sqlite3.connect(":memory:")

    return conn
```

```python
con =InitalSqliteConnection('chinaaid.db')
con.text_factory = str
ca=pd.read_csv('ChinaAid_own.csv')
ca.to_sql('ChinaAidProjectlist',con,if_exists='replace')
```

```python
if __name__ == '__main__':
    requests_cache.install_cache(
        cache_name="ChinaAidProjectInfo", backend="sqlite", expire_date=300)
    r = requests.get(
        'http://china.aiddata.org/aggregates/projects?get=donor,year')
    
    rr = pd.DataFrame(eval(r.text))



    

    
```

```python
param =['donor'
,'status'
,'intent'
,'year'
,'crs_sector_name'
,'flow_class'
,'recipient_name'
,'recipient_iso2'
,'recipient_iso3']
r2 = requests.get(
        'http://china.aiddata.org/aggregates/projects?get=%s'%(','.join(param)))
#print ','.join(param)

recipent =pd.DataFrame(eval(r2.text.replace('null','"NULL"')))
```

```python
recipent.to_hdf('ChinaAid.h5','ChinaAid',complevel =9,complib='blosc')
```

```python
yearandrecipent_based=[]
for year,year_df in recipent.groupby('year'):
    yearandrecipent_based.extend([(year,recipients,recipients_df.usd_current.sum()) for recipients,recipients_df in year_df.groupby('recipient_name')])
         
pd.DataFrame(yearandrecipent_based,columns=['year','recipient_name','totalamount'])      
```
